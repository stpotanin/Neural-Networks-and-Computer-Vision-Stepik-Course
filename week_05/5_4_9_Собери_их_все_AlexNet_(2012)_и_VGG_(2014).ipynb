```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Определение класса для свёртки с расширением
class DilatedConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, dilation=1):
        super(DilatedConv2d, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, padding=dilation, dilation=dilation)

    def forward(self, x):
        return self.conv(x)

# Создание входного изображения
input_image = torch.randn(1, 1, 3, 3)

# Создание свёрточных слоёв с различными расширениями
conv_0 = DilatedConv2d(1, 1, 3, dilation=1)
conv_1 = DilatedConv2d(1, 1, 3, dilation=2)
conv_2 = DilatedConv2d(1, 1, 3, dilation=3)

# Применение свёрток к входному изображению
output_0 = conv_0(input_image)
output_1 = conv_1(input_image)
output_2 = conv_2(input_image)

# Визуализация входного изображения и выходов
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.imshow(input_image[0, 0, :, :].detach().numpy(), cmap='gray')
plt.title('Input Image')

plt.subplot(2, 2, 2)
plt.imshow(output_0[0, 0, :, :].detach().numpy(), cmap='gray')
plt.title('Output (dilation=1)')

plt.subplot(2, 2, 3)
plt.imshow(output_1[0, 0, :, :].detach().numpy(), cmap='gray')
plt.title('Output (dilation=2)')

plt.subplot(2, 2, 4)
plt.imshow(output_2[0, 0, :, :].detach().numpy(), cmap='gray')
plt.title('Output (dilation=3)')

plt.tight_layout()
plt.show()
```
