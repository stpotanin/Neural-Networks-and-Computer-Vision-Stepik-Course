У нас есть задача минимизации суммы квадратов отклонений между некоторым средним значением $\tilde{y}$ и наблюдениями $\{y_i\}_{i=1}^N$. Формально это записано как:

$$
μ = \underset{\tilde{y}}{\arg \min} \sum_{i=1}^N (\tilde{y} - y_i)^2
$$

1. **Постановка задачи:** Нужно найти такое значение $\tilde{y}$, которое минимизирует функцию потерь $L(\tilde{y}) = \sum_{i=1}^N (\tilde{y} - y_i)^2$. Эта функция потерь описывает сумму квадратов отклонений $\tilde{y}$ от наблюдений $y_i$.

2. **Нахождение производной функции потерь:** Чтобы найти минимум функции, нам нужно взять производную от функции потерь по переменной $\tilde{y}$ и приравнять её к нулю:

$$
\frac{\partial L}{\partial \tilde{y}} = \frac{\partial}{\partial \tilde{y}} \sum_{i=1}^N (\tilde{y} - y_i)^2
$$

3. **Дифференцирование суммы:** Каждый элемент суммы $(\tilde{y} - y_i)^2$ можно дифференцировать отдельно:

$$
\frac{\partial}{\partial \tilde{y}} (\tilde{y} - y_i)^2 = 2(\tilde{y} - y_i)
$$

Поэтому производная всей суммы будет равна:

$$
\frac{\partial L}{\partial \tilde{y}} = \sum_{i=1}^N 2(\tilde{y} - y_i) = 2\sum_{i=1}^N (\tilde{y} - y_i)
$$

4. **Приравнивание к нулю:** Чтобы найти минимум, приравниваем производную к нулю:

$$
2\sum_{i=1}^N (\tilde{y} - y_i) = 0
$$

5. **Решение уравнения:** Раскрываем сумму:

$$
2N\tilde{y} - 2\sum_{i=1}^N y_i = 0
$$

Убираем множитель $2$:

$$
N\tilde{y} = \sum_{i=1}^N y_i
$$

Отсюда:

$$
\tilde{y} = \frac{1}{N} \sum_{i=1}^N y_i
$$

Таким образом, $ \tilde{y} $ — это просто среднее арифметическое наблюдений $y_i$, то есть:

$$
μ = \frac{1}{N} \sum_{i=1}^N y_i
$$

Это означает, что решение задачи минимизации суммы квадратов отклонений — это среднее арифметическое всех наблюдений.
