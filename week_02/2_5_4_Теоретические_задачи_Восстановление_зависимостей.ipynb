# Решение
Если все элементы набора $ x_1, x_2, \dots, x_N $ **уникальные**, то матрица Вандермонда $ X $ будет иметь полный ранг при определённых условиях. Для того чтобы уравнение $ y = Xa $ имело **единственное решение**, должны выполняться следующие условия:

1. **Полнота ранга матрицы**: Для уникальности решения важно, чтобы матрица Вандермонда была полноранговой. В случае, когда все элементы $ x_i $ уникальны, матрица Вандермонда будет полноранговой, если количество столбцов $ M $ не превышает количество строк $ N $, то есть:
   $$
   M \leq N
   $$
   Это связано с тем, что при $ M \leq N $ матрица Вандермонда будет иметь полный ранг $ M $, и система уравнений $ y = Xa $ будет определена однозначно.

2. **Условие для существования единственного решения**: Чтобы существовало единственное решение, количество столбцов $ M $ должно точно соответствовать количеству неизвестных параметров $ a $. Если $ M = N $, система становится квадратной, и при уникальных $ x_i $ матрица Вандермонда будет невырожденной, обеспечивая единственное решение.

### Итог:
Для уникальности решения уравнения $ y = Xa $ при уникальных $ x_i $, должно выполняться условие:
$$
M = N
$$
В этом случае система будет квадратной, матрица Вандермонда будет невырожденной, и уравнение будет иметь единственное решение.

Если $ M < N $, возможно найти множество решений (если система переопределена), а если $ M > N $, система будет недоопределенной, и решений либо не будет, либо будет бесконечно много.

Решение оптимизационной задачи и решение системы уравнений — это не всегда одно и то же.
Когда мы говорим о решении системы уравнений $y=Xa$, мы имеем в виду точное решение, при котором каждый элемент вектора $y$ точно равен произведению матрицы $X$ на вектор коэффициентов $a$.
Когда система уравнений переопределена $(N>M)$, невозможно найти точное решение для всех уравнений одновременно. В этом случае мы можем сформулировать оптимизационную задачу, где минимизируется ошибка (расхождение) между тем, что предсказывает система, и наблюдаемыми значениями $y$.

$$
a=arg _a min∥Xa-y∥
2
$$

Это не обязательно даст точное решение для каждого уравнения, но позволит найти наилучшее приближение, минимизируя ошибку.

В этом случае, хотя количество уравнений больше, чем количество переменных, система имеет единственное решение в смысле наименьших квадратов.

# Пример
Пример системы с матрицей Вандермонда, где $ M < N $ (то есть она "высокая") и существует **ровно одно решение**, — это задача на **метод наименьших квадратов** (МНК). В таких случаях система уравнений может быть переопределённой (больше уравнений, чем переменных), но можно найти единственное решение через минимизацию ошибки.

### Пример:
Рассмотрим следующую задачу:
- Пусть $ N = 3 $ (три уравнения),
- Пусть $ M = 2 $ (две переменные $ a_0 $ и $ a_1 $).

Значения $ x_1 = 1 $, $ x_2 = 2 $, $ x_3 = 3 $ — все уникальные.

Тогда матрица Вандермонда $ X $ для полинома степени 1 будет выглядеть так:
$$
X = \begin{pmatrix}
1 & 1 \\
1 & 2 \\
1 & 3
\end{pmatrix}
$$
Пусть вектор наблюдений $ y = \begin{pmatrix} 2 \\ 3 \\ 5 \end{pmatrix} $.

Мы хотим решить систему $ y = Xa $, где $ a = \begin{pmatrix} a_0 \\ a_1 \end{pmatrix} $ — вектор коэффициентов, которые мы ищем.

### Решение через метод наименьших квадратов:

Так как у нас больше уравнений, чем переменных ($ N = 3 $, $ M = 2 $), система **переопределена**, но единственное решение можно найти через минимизацию квадратичной ошибки, решая задачу:
$$
a = \underset{a}{\arg\min} \|Xa - y\|^2
$$
Для нахождения этого решения мы используем нормальное уравнение:
$$
a = (X^T X)^{-1} X^T y
$$

### Пояснение:
- В этом случае, хотя количество уравнений больше, чем количество переменных, система имеет **единственное решение** в смысле наименьших квадратов. Матрица Вандермонда здесь имеет полный ранг, и единственное решение будет найдено через минимизацию ошибки, а не точное совпадение всех уравнений.
