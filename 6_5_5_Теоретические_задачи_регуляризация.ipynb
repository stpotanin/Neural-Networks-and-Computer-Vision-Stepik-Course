# Решение
Рассмотрим задачу оптимизации:

$$
\frac{1}{2}(wx - b)^2 + \lambda |w| \to \min_w,
$$

где $ x $, $ b $ — действительные числа, а $ \lambda \geq 0 $.

### Шаг 1: Производная функции
Чтобы найти точку минимума, сначала возьмём производную функции по $ w $. При этом учтем, что $ |w| $ является разрывной функцией, поэтому нужно рассматривать два случая: $ w > 0 $ и $ w < 0 $.

#### 1. Случай $ w > 0 $:

Функция выглядит так:

$$
f(w) = \frac{1}{2}(wx - b)^2 + \lambda w.
$$

Найдём её производную:

$$
f'(w) = (wx - b) \cdot x + \lambda = x(wx - b) + \lambda.
$$

Решаем уравнение $ f'(w) = 0 $:

$$
x(wx - b) + \lambda = 0,
$$
$$
x^2 w - bx + \lambda = 0,
$$
$$
w = \frac{b x - \lambda}{x^2}.
$$

#### 2. Случай $ w < 0 $:

В этом случае функция принимает вид:

$$
f(w) = \frac{1}{2}(wx - b)^2 - \lambda w.
$$

Найдём её производную:

$$
f'(w) = (wx - b) \cdot x - \lambda = x(wx - b) - \lambda.
$$

Решаем уравнение $ f'(w) = 0 $:

$$
x(wx - b) - \lambda = 0,
$$
$$
x^2 w - bx - \lambda = 0,
$$
$$
w = \frac{b x + \lambda}{x^2}.
$$

### Шаг 2: Когда $ w = 0 $ является решением

Рассмотрим случай, когда $ w = 0 $. Подставляем $ w = 0 $ в исходное уравнение:

$$
f(0) = \frac{1}{2}(0 \cdot x - b)^2 + \lambda \cdot 0 = \frac{1}{2}b^2.
$$

Теперь посмотрим, при каком значении $ \lambda $ будет выгоднее выбирать $ w = 0 $, чем любое другое $ w \neq 0 $.

Для $ w > 0 $:

$$
w = \frac{b x - \lambda}{x^2}.
$$

Для того чтобы $ w \neq 0 $, должно выполняться условие:

$$
b x - \lambda > 0 \quad \Rightarrow \quad \lambda < b x.
$$

Таким образом, если $ \lambda \geq |bx| $, то оптимальным решением будет $ w = 0 $.

### Ответ:
Задача имеет решение $ w \neq 0 $, если $ \lambda < |bx| $.

### Мой Шаг 2

$$
w = \frac{b x + \lambda}{x^2}.
$$

Из условия $\lambda$ всегда положительна (точнее, неотрицательна).

А вот произведение  $b x$ в полученных выше формулах может быть как положительным, так и отрицательным.

Если оно положительно, то в раскладе
$$
w = \frac{b x + \lambda}{x^2}.
$$
об ограничении $\lambda$ беспокоиться почти не надо (если только произведение не выродитсся в ноль).

А вот при отрицательном (и нулевом) $b x$ нужно, чтобы $\lambda$ была меньше модуля этого произведения: $ \lambda < |bx| $.

Во втором раскладе, где
$$
w = \frac{b x - \lambda}{x^2}.
$$
всё наоборот.

При отрицательном $b x$ беспокоиться не надо, знаменатель гарантированно уйдёт в минус.

А вот при положительном $b x$, опять-таки необходимо, чтобы $\lambda$ была меньше модуля произведения $b x$: $ \lambda < |bx| $.

В итоге - один ответ для обоих раскладов.
