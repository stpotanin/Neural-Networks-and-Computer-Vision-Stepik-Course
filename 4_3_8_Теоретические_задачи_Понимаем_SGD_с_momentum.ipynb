# Решение
### Пояснение

1. **Взвешенное среднее:**

   Взвешенное среднее некоторой величины $ f $ определяется формулой:
   $$
   \bar{f} = \frac{\sum_{i=1}^{N} f_i w_i}{\sum_{i=1}^{N} w_i}
   $$
   где $ w_i $ — вес, с которым взвешивается каждое значение $ f_i $.

2. **Экспоненциальное скользящее среднее (EMA):**

   Экспоненциальное скользящее среднее вычисляется рекурсивно:
   $$
   EMA_{\alpha} f_{t+1} = \alpha \cdot EMA_{\alpha} f_t + (1 - \alpha) \cdot f_{t+1}
   $$
   где $ \alpha $ — коэффициент сглаживания, а начальное значение $ EMA_{\alpha} f_0 = 0 $.

   Можно показать, что при $ t \to \infty $ EMA стремится к взвешенному среднему, но при малых $ t $ это не так.

### Суть задачи:

Экспоненциальное скользящее среднее на самом деле является взвешенным средним, но вес для каждого наблюдения накапливается со временем и экспоненциально уменьшается. Чтобы получить взвешенное среднее на любом шаге $ t $, нужно корректировать $ EMA_{\alpha} f_t $ соответствующим весовым коэффициентом.

### Решение:

Экспоненциальное среднее можно представить в виде взвешенного среднего:

$$
EMA_{\alpha} f_t = \frac{\sum_{i=1}^{t} f_i \cdot w_i}{\sum_{i=1}^{t} w_i}
$$

Здесь веса $ w_i = (1-\alpha) \cdot \alpha^{t-i} $, а нормировка $\sum_{i=1}^{t} w_i$ позволяет получить корректное взвешенное среднее.

Следовательно, для получения взвешенного среднего нужно разделить экспоненциальное среднее на суммарный вес:

$$
\bar{f_t} = \frac{EMA_{\alpha} f_t}{\sum_{i=1}^{t} (1-\alpha) \cdot \alpha^{t-i}}
$$

Или что эквивалентно:

$$
\bar{f_t} = \frac{EMA_{\alpha} f_t}{1 - \alpha^t}
$$

**Ответ:** Чтобы экспоненциальное скользящее среднее $ EMA_{\alpha} f_t $ стало взвешенным средним, его нужно разделить на $ 1 - \alpha^t $.
