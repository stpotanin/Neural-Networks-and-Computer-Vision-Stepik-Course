# Ядро матрицы (также применимо к оператору или пространству)
— это множество всех векторов, которые при умножении на эту матрицу дают нулевой вектор. В терминах линейного преобразования, ядро представляет собой множество всех входных векторов, которые преобразуются в нулевой вектор на выходе.

Формально, для матрицы $ A $, ядро определяется как:

$$ \text{Ядро}(A) = \{ \vec{x} \mid A \vec{x} = \vec{0} \} $$

Это множество векторов, которые смещаются в нулевой вектор под действием матрицы $ A $.

Базис матрицы (точнее, базис пространства строк матрицы) и базис её ядра ортогональны друг другу.
