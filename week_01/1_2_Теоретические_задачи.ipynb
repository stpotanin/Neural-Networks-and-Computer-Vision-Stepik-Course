2. –ù–∞–π–¥–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $ \frac{dy}{dx} $:

$ \frac{du}{dx} = \frac{d}{dx} (1 + e^{-x}) = -e^{-x} $

# –†–µ—à–µ–Ω–∏–µ
–§—É–Ω–∫—Ü–∏—è –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞ (tanh) —è–≤–ª—è–µ—Ç—Å—è —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω—ë–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ –≤ –Ω–µ–π—Ä–æ–Ω–Ω—ã—Ö —Å–µ—Ç—è—Ö. –ï—ë –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

$ \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–≤–æ–π—Å—Ç–≤–∞ —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

### 1. –ß–µ—Ç–Ω–æ—Å—Ç—å
–§—É–Ω–∫—Ü–∏—è –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞ –Ω–µ—á—ë—Ç–Ω–∞—è. –≠—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ $\tanh(-x) = -\tanh(x)$. –î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, –µ—Å–ª–∏ –ø–æ–¥—Å—Ç–∞–≤–∏—Ç—å $-x$ –≤ —Ñ—É–Ω–∫—Ü–∏—é:

$ \tanh(-x) = \frac{e^{-x} - e^{x}}{e^{-x} + e^{x}} = -\frac{e^x - e^{-x}}{e^x + e^{-x}} = -\tanh(x) $

### 2. –ú–æ–Ω–æ—Ç–æ–Ω–Ω–æ—Å—Ç—å –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è
–§—É–Ω–∫—Ü–∏—è $\tanh(x)$ —è–≤–ª—è–µ—Ç—Å—è –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â–µ–π. –ï—ë –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è:

$ \frac{d}{dx} \tanh(x) = 1 - \tanh^2(x) $

–í—Å–µ–≥–¥–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞ –¥–ª—è –≤—Å–µ—Ö $x$, —Ç–∞–∫ –∫–∞–∫ $ 1 - \tanh^2(x) \geq 0 $ –¥–ª—è –≤—Å–µ—Ö $x$.

### 3. –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å
–§—É–Ω–∫—Ü–∏—è –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –¥–ª—è –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π $x$. –≠—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –¥–ª—è –ª—é–±–æ–≥–æ $a$ –∏–∑ –æ–±–ª–∞—Å—Ç–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–µ–¥–µ–ª $\tanh(x)$ –ø—Ä–∏ $x \to a$ —Ä–∞–≤–µ–Ω $\tanh(a)$:

$ \lim_{x \to a} \tanh(x) = \tanh(a) $

### 4. –û–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å
–§—É–Ω–∫—Ü–∏—è $\tanh(x)$ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∞. –ï—ë –∑–Ω–∞—á–µ–Ω–∏—è –ª–µ–∂–∞—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ –æ—Ç -1 –¥–æ 1:

$ -1 \leq \tanh(x) \leq 1 $

### 5. –†–∞–∑—Ä—ã–≤—ã
–§—É–Ω–∫—Ü–∏—è $\tanh(x)$ –Ω–µ –∏–º–µ–µ—Ç —Ä–∞–∑—Ä—ã–≤–æ–≤. –û–Ω–∞ –≥–ª–∞–¥–∫–∞—è –∏ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è –¥–ª—è –≤—Å–µ—Ö $x$.

### –†–µ–∑—é–º–µ

- **–ú–æ–Ω–æ—Ç–æ–Ω–Ω–æ—Å—Ç—å –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è**: –ú–æ–Ω–æ—Ç–æ–Ω–Ω–æ –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â–∞—è.
- **–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å**: –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è –¥–ª—è –≤—Å–µ—Ö $x$.
- **–û–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å**: –ó–Ω–∞—á–µ–Ω–∏—è –æ–≥—Ä–∞–Ω–∏—á–µ–Ω—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º $[-1, 1]$.
- **–†–∞–∑—Ä—ã–≤—ã**: –ù–µ—Ç —Ä–∞–∑—Ä—ã–≤–æ–≤.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–∏–π —Ç–∞–Ω–≥–µ–Ω—Å $\tanh(x)$ –æ–±–ª–∞–¥–∞–µ—Ç –≤–∞–∂–Ω—ã–º–∏ —Å–≤–æ–π—Å—Ç–≤–∞–º–∏, –∫–æ—Ç–æ—Ä—ã–µ –¥–µ–ª–∞—é—Ç –µ–≥–æ –ø–æ–ª–µ–∑–Ω—ã–º –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ –≤ –Ω–µ–π—Ä–æ–Ω–Ω—ã—Ö —Å–µ—Ç—è—Ö.

```python
import numpy as np
import matplotlib.pyplot as plt

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞
def tanh(x):
    return np.tanh(x)

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞
# def tanh_derivative(x):
#     return 1 - np.tanh(x)**2

# –¢–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
x = np.linspace(-5, 5, 400)
y = tanh(x)
# y_prime = tanh_derivative(x)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞ –∏ –µ—ë –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π
plt.figure(figsize=(10, 5))

# –ì–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–∏–π —Ç–∞–Ω–≥–µ–Ω—Å
plt.subplot(1, 2, 1)
plt.plot(x, y, label='tanh(x)')
plt.title('Hyperbolic Tangent Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()

# –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –≥–∏–ø–µ—Ä–±–æ–ª–∏—á–µ—Å–∫–æ–≥–æ —Ç–∞–Ω–≥–µ–Ω—Å–∞
# plt.subplot(1, 2, 2)
# plt.plot(x, y_prime, label="tanh'(x)", color='red')
# plt.title('Derivative of Hyperbolic Tangent Function')
# plt.xlabel('x')
# plt.ylabel("tanh'(x)")
# plt.grid(True)
# plt.legend()

# plt.tight_layout()
# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y = np.tanh(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Hyperbolic Tangent Function')
plt.grid(True)
plt.show()
```
