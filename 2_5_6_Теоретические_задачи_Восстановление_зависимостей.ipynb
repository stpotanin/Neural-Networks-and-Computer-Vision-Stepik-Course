# Решение OpenAI

Рассмотрим задачу оптимизации $ L(a) = \|Xa - y\|_2^2 $, где $ X $ — матрица размером $ M \times N $, $ a $ — вектор размером $ N \times 1 $, а $ y $ — вектор размером $ M \times 1 $. При условии, что $ M > N $, решение этой задачи не единственно, поскольку система уравнений переопределена.

Чтобы найти матрицу, собственные векторы которой, соответствующие нулевым собственным числам, образуют ядро $ X $, нужно рассмотреть матрицу $ X $ и ее свойства.

Ядро (нулевое пространство) матрицы $ X $ — это множество всех векторов $ a $, таких что $ Xa = 0 $. Иными словами, мы ищем матрицу $ A $, чьи собственные векторы, соответствующие нулевым собственным числам, составляют это пространство.

Матрица $ A $, которая нам нужна, — это матрица $ X^T X $, где $ X^T $ — транспонированная матрица $ X $. Давайте объясним, почему это так:

1. **Собственные числа $ X^T X $**: Если $ \lambda $ — собственное число матрицы $ X^T X $, а $ v $ — соответствующий собственный вектор, то:
   $$
   X^T X v = \lambda v
   $$
   Для $ v \neq 0 $, если $ \lambda = 0 $, это означает, что:
   $$
   X^T X v = 0 \implies X v = 0
   $$
   То есть $ v $ находится в ядре матрицы $ X $.

2. **Собственные векторы**: Собственные векторы, соответствующие нулевым собственным числам матрицы $ X^T X $, являются векторами из ядра $ X $. Эти векторы составляют базис ядра матрицы $ X $.

Итак, матрица $ X^T X $ имеет те же собственные векторы, что и $ X $, и нулевые собственные числа этой матрицы дают нам ядро $ X $.

### Пример:
Рассмотрим конкретный пример для иллюстрации:

Пусть
$$ X = \begin{pmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{pmatrix} $$
Тогда
$$ X^T = \begin{pmatrix}
1 & 3 & 5 \\
2 & 4 & 6
\end{pmatrix} $$
и
$$ X^T X = \begin{pmatrix}
1 & 3 & 5 \\
2 & 4 & 6
\end{pmatrix}
\begin{pmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{pmatrix} = \begin{pmatrix}
35 & 44 \\
44 & 56
\end{pmatrix} $$

Собственные числа матрицы $ X^T X $ равны примерно 0.26 и 90.74 (вы можете проверить это, решив характеристическое уравнение). Собственный вектор, соответствующий нулевому собственному числу, дает нам ядро $ X $.

Таким образом, матрица $ X^T X $ является искомой матрицей, чьи собственные векторы, соответствующие нулевым собственным числам, образуют ядро $ X $.

```python
# Проверка
from numpy import array
import numpy as np

X = array([[1, 2], [3, 4], [5, 6]])


def eigenvalues_eigenvectors(X):
    if X.shape[0] != X.shape[1]:
        print("Ошибка: Матрица должна быть квадратной для вычисления собственных значений.")
    else:
        eigenvalues, eigenvectors = np.linalg.eig(X)

        print("Собственные числа:\n", eigenvalues)
        print("Собственные векторы:\n", eigenvectors)

eigenvalues_eigenvectors(X.T.dot(X))
```
