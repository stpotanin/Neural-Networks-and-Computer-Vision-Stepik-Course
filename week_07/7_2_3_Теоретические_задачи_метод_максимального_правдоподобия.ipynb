Задача заключается в нахождении вероятности зарегистрировать $ M $ фотонов на фоточувствительной пластине в условиях, когда:

1. Источник генерирует $ N $ фотонов в единицу времени.
2. Каждый фотон с вероятностью $ p_s $ может рассеяться в среде и не достичь пластины.
3. Каждый фотон, достигший пластины, с вероятностью $ p_d $ может быть зарегистрирован.

```python
# C(M,N)*((1-p_s)*p_d)^M*(1-(1-p_s)*p_d)^(N-M)
```

# Решение

### Рассмотрим процесс по шагам:

#### 1. Вероятность того, что фотон достигнет пластины.
Пусть вероятность того, что фотон не рассеивается и достигнет пластины, равна $ 1 - p_s $. Тогда для каждого фотона вероятность достичь пластины будет:

$$
P_{\text{reach}} = 1 - p_s.
$$

#### 2. Вероятность того, что фотон будет зарегистрирован пластиной.
Если фотон достиг пластины, то вероятность его регистрации равна $ p_d $. Тогда общая вероятность того, что фотон будет зарегистрирован (т.е., он не рассеялся и был зарегистрирован) равна:

$$
P_{\text{registered}} = P_{\text{reach}} \cdot p_d = (1 - p_s) p_d.
$$

#### 3. Вероятность зарегистрировать $ M $ фотонов.
Зарегистрированные фотоны можно рассматривать как результат независимых испытаний, где каждый фотон регистрируется с вероятностью $ (1 - p_s) p_d $. Это биномиальное распределение с параметрами $ N $ (количество испытаний — фотонов) и вероятностью успеха $ P_{\text{registered}} = (1 - p_s) p_d $.

Вероятность того, что будет зарегистрировано ровно $ M $ фотонов, описывается биномиальным распределением:

$$
P(M) = C(M, N) \cdot \left( (1 - p_s) p_d \right)^M \cdot \left( 1 - (1 - p_s) p_d \right)^{N - M}.
$$

Здесь:
- $ C(M, N) $ — биномиальный коэффициент, который выражает количество способов выбрать $ M $ зарегистрированных фотонов из $ N $ фотонов, то есть:

$$
C(M, N) = \frac{N!}{M!(N - M)!}.
$$

- $ \left( (1 - p_s) p_d \right)^M $ — вероятность зарегистрировать ровно $ M $ фотонов.
- $ \left( 1 - (1 - p_s) p_d \right)^{N - M} $ — вероятность того, что остальные $ N - M $ фотонов не будут зарегистрированы.

### Итоговая формула:
Вероятность зарегистрировать $ M $ фотонов на фоточувствительной пластине равна:

$$
P(M) = C(M, N) \cdot \left( (1 - p_s) p_d \right)^M \cdot \left( 1 - (1 - p_s) p_d \right)^{N - M}.
$$

### Биномиальное распределение

Биномиальное распределение описывает количество успехов в серии независимых испытаний, где каждое испытание может завершиться либо успехом, либо неудачей.

### Параметры биномиального распределения:

- $ N $ — количество испытаний (в нашей задаче это количество фотонов, $ N $).
- $ p $ — вероятность успеха в каждом испытании (в нашем случае это вероятность зарегистрировать фотон, которая равна $ P_{\text{registered}} = (1 - p_s)p_d $).

Если каждое испытание может завершиться успехом с вероятностью $ p $, то биномиальное распределение задается вероятностью того, что в $ N $ испытаниях будет ровно $ M $ успехов.

#### Формула биномиального распределения:

Вероятность того, что в $ N $ независимых испытаниях произойдет ровно $ M $ успехов, равна:

$$
P(M) = C(M, N) \cdot p^M \cdot (1 - p)^{N - M},
$$

где:
- $ C(M, N) = \frac{N!}{M!(N - M)!} $ — биномиальный коэффициент (он задает число способов выбрать $ M $ успешных испытаний из $ N $).
- $ p^M $ — вероятность того, что $ M $ испытаний завершатся успехом.
- $ (1 - p)^{N - M} $ — вероятность того, что оставшиеся $ N - M $ испытаний завершатся неудачей.

### Пример в контексте задачи

Для нашей задачи:

- $ N $ — это количество фотонов, которые излучает источник.
- $ p = P_{\text{registered}} = (1 - p_s)p_d $ — вероятность того, что фотон достигнет пластины и будет зарегистрирован (успех).
- $ M $ — это количество зарегистрированных фотонов.

Вероятность зарегистрировать ровно $ M $ фотонов на пластине описывается биномиальным распределением:

$$
P(M) = C(M, N) \cdot \left( (1 - p_s)p_d \right)^M \cdot \left( 1 - (1 - p_s)p_d \right)^{N - M}.
$$

#### Особенности этого распределения:

1. **Ожидаемое значение (математическое ожидание)**. Среднее количество зарегистрированных фотонов $ \mathbb{E}(M) $ (ожидаемое значение) для биномиального распределения определяется как:

$$
\mathbb{E}(M) = N \cdot p = N \cdot (1 - p_s) \cdot p_d.
$$

2. **Дисперсия**. Разброс количества зарегистрированных фотонов (дисперсия) определяется формулой:

$$
\text{Var}(M) = N \cdot p \cdot (1 - p) = N \cdot (1 - p_s) \cdot p_d \cdot \left(1 - (1 - p_s) p_d \right).
$$

### Интерпретация:

- Чем больше $ p_d $ (вероятность регистрации фотона на пластине), тем выше вероятность того, что будет зарегистрировано больше фотонов.
- Чем больше $ p_s $ (вероятность рассеяния фотонов), тем меньше вероятность регистрации фотонов, так как большее их количество не достигнет пластины.
- Биномиальное распределение описывает вероятностное поведение системы с независимыми событиями (фотоны достигают или не достигают пластины независимо друг от друга).
