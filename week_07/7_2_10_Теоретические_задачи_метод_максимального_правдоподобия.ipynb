# Условие

Допустим, что мы знаем $p(x|λ)$ -- вероятность наблюдать $x$ при условии параметра $λ$. Допустим, что мы также откуда-то знаем вероятности различных значений $λ:p(λ)$.

Найдите логарифм вероятности $p(λ|x)$. Отбросьте все члены, которые не зависят от $λ$.

Для этого вам может понадобится [формула Байеса](https://ru.wikipedia.org/wiki/%D0%A2%D0%B5%D0%BE%D1%80%D0%B5%D0%BC%D0%B0_%D0%91%D0%B0%D0%B9%D0%B5%D1%81%D0%B0)

$λ$ обозначайте как `l`.

# Решение

**Формула Байеса:**

$$
p(\lambda|x) = \frac{p(x|\lambda) p(\lambda)}{p(x)}
$$

где:

* $p(\lambda|x)$ - апостериорная вероятность $\lambda$ при условии наблюдения $x$.
* $p(x|\lambda)$ - правдоподобие, вероятность наблюдать $x$ при условии $\lambda$.
* $p(\lambda)$ - априорная вероятность $\lambda$.
* $p(x)$ - маргинальная вероятность $x$ (вероятность наблюдать $x$ независимо от $\lambda$).


**Логарифм апостериорной вероятности:**

Возьмем логарифм от обеих сторон формулы Байеса:

$$
\log p(\lambda|x) = \log \left[ \frac{p(x|\lambda) p(\lambda)}{p(x)} \right]
$$

$$
\log p(\lambda|x) = \log p(x|\lambda) + \log p(\lambda) - \log p(x)
$$


**Отбрасывание членов, не зависящих от $\lambda$:**

В выражении выше, только члены $\log p(x|\lambda)$ и $\log p(\lambda)$ зависят от $\lambda$. Член $\log p(x)$ не зависит от $\lambda$, так как $p(x)$ - это маргинальная вероятность $x$, которая рассчитывается путем интегрирования (суммирования) по всем возможным значениям $\lambda$.

**Результат:**

Таким образом, отбросив члены, не зависящие от $\lambda$, получаем:

$$
\log p(\lambda|x) \propto \log p(x|\lambda) + \log p(\lambda)
$$


**Вывод:**

Логарифм апостериорной вероятности $p(\lambda|x)$, если отбросить члены, не зависящие от $\lambda$, пропорционален сумме логарифма правдоподобия $p(x|\lambda)$ и логарифма априорной вероятности $p(\lambda)$.
